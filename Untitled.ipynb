```python
credit_data["Annual_Income"] = credit_data["Annual_Income"].str.replace("_","")
```

```python
occupation_group = credit_data.groupby("Occupation").aggregate({"Annual_Income":"mean"})
occupation_group.sort_values(by="Annual_Income",ascending=False)

plt.figure(figsize=(8, 6))
sns.barplot(x=occupation_group.index, y='Annual_Income', data=occupation_group.reset_index())
plt.title('Occupation vs. Average Annual Income')
plt.xlabel('Occupation')
plt.ylabel('Average Annual Income')
plt.xticks(rotation=45, ha='right')
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(x='Occupation', y='Annual_Income', data=credit_data)
plt.title('Occupation vs. Annual Income Distribution')
plt.xlabel('Occupation')
plt.ylabel('Annual Income')
plt.xticks(rotation=45, ha='right')
plt.show()
```
